# Plot Correlation Heatmap

```python
#Load required dependencies
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
def make_heatmap(name):
    base_path = "/Users/emmacorcoran/Documents/R Scripts/RNA-seq/2021_09_16 RNA-seq/Correlation/"
    save_path = "/Users/emmacorcoran/Documents/Jupyter Code/RNA-seq"
    path_to_file = os.path.join(base_path, name)
    data_to_plot = pd.read_csv(path_to_file, index_col=0, sep='\t')
    #rocket scale
    sns.set(font_scale = 1.5);
    g = sns.clustermap(data_to_plot, cmap='rocket_r', annot=True);
    save_name = name.split(".")[0] + "_rocket.png"
    g.savefig(os.path.join(save_path, save_name))
    #Blues scale
    sns.set(font_scale = 1.5);
    g = sns.clustermap(data_to_plot, cmap='Blues', annot=True);
    save_name = name.split(".")[0] + "_Blues.png"
    g.savefig(os.path.join(save_path, save_name))
```

```python
directory = "/Users/emmacorcoran/Documents/R Scripts/RNA-seq/2021_09_16 RNA-seq/Correlation/"
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        make_heatmap(filename)
```
